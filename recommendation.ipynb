## Prepare data for analysis and model building

For preprocessing, we will do the following

1) Join the dataset using the ID variable 

2) Preprocess the joined dataset






```python
credits_rename = credits.rename(columns={'movie_id':'id'})
```

```python
credits_rename
```

```python
movies_df_merge = movies.merge(credits_rename, on = 'id')
movies_df_merge.head()
```

```python
#Get the shape of the merge dataset
movies_df_merge.shape
```

```python
#Get all features of the dataset
movies_df_merge.columns
```

```python
# Exclude un important features from our dataset
movies_clean_df = movies_df_merge.drop(columns=['title_y', 'title_x', 'homepage','production_countries', 'status' ])
```
